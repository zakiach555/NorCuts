# 🎬 NorCuts - AI-Powered Video Clip Generator

**A free, open-source alternative to Opus Clip and Vidyo.ai**

Transform long YouTube videos into viral short clips optimized for TikTok, Instagram Reels, and YouTube Shorts using state-of-the-art AI.

## Features ✨
- 🤖 **AI Viral Detection**: Automatically identifies engaging moments using Gemini, GPT-4, or Local LLMs
- 🗣️ **Ultra-Precise Transcription**: Powered by WhisperX with GPU acceleration
- 🎨 **Dynamic Captions**: Hormozi-style with word-by-word highlights and customization
- 🎥 **Smart Face Tracking**: Auto-crop 9:16, split-screen for 2 faces, active speaker detection
- 🌍 **Translation Support**: Translate captions to multiple languages
- 💾 **Professional Export**: Up to 4K resolution, XML export to Premiere Pro

## 🛠️ Step 1: Installation

Run this cell to install all dependencies. This will take approximately 5-10 minutes.

```python
# @title 🚀 Install NorCuts Dependencies
import os
import subprocess
import shutil
from IPython.display import clear_output

# 1. Clean previous installation
print("🧹 Cleaning previous installation...")
%cd /content
if os.path.exists("NorCuts"):
    shutil.rmtree("NorCuts")

# Clone the repository
print("📦 Cloning NorCuts repository...")
!git clone https://github.com/RafaelGodoyEbert/ViralCutter.git NorCuts
%cd /content/NorCuts

# 2. Install UV package manager and system drivers
print("⏳ Installing UV package manager and system dependencies...")
subprocess.run(['pip', 'install', 'uv'], check=True, capture_output=True)
subprocess.run('sudo apt update -y && sudo apt install -y libcudnn8 ffmpeg xvfb', shell=True, check=True, capture_output=True)

# 3. Create virtual environment
print("⏳ Creating virtual environment...")
subprocess.run(['uv', 'venv', '.venv'], check=True, capture_output=True)

# 4. Install dependencies
print("⏳ Installing Python libraries (this may take a few minutes)...")

# Phase A: Core dependencies
cmds_phase_1 = [
    "uv pip install --python .venv git+https://github.com/m-bain/whisperx.git",
    "uv pip install --python .venv -r requirements.txt",
    "uv pip install --python .venv yt-dlp pytubefix",
    "uv pip install --python .venv yt_dlp"
]

for cmd in cmds_phase_1:
    subprocess.run(cmd, shell=True, check=True, capture_output=True)

# Phase B: AI and alignment fixes
print("🔨 Applying strategic downgrades for compatibility...")
extra_libs = [
    "uv pip install --python .venv google-generativeai",
    "uv pip install --python .venv pandas",
    "uv pip install --python .venv onnxruntime-gpu",
    "uv pip install --python .venv transformers==4.46.3 accelerate>=0.26.0"
]

for cmd in extra_libs:
    subprocess.run(cmd, shell=True, check=True, capture_output=True)

# Phase C: Force stable PyTorch version
print("🔨 Forcing stable PyTorch version (2.3.1)...")
cmd_fix_torch = (
    "uv pip install --python .venv "
    "torch==2.3.1+cu121 torchvision==0.18.1+cu121 torchaudio==2.3.1+cu121 "
    "--index-url https://download.pytorch.org/whl/cu121"
)
subprocess.run(cmd_fix_torch, shell=True, check=True, capture_output=True)

# Phase D: Lock NumPy version
print("🔨 Locking NumPy version...")
subprocess.run("uv pip install --python .venv 'numpy<2.0' setuptools==69.5.1", shell=True, check=True, capture_output=True)

# Phase E: Fix computer vision dependencies
print("🔨 Fixing InsightFace and MediaPipe...")
subprocess.run("uv pip install --python .venv insightface onnxruntime-gpu", shell=True, check=True, capture_output=True)
subprocess.run("uv pip uninstall --python .venv mediapipe protobuf flatbuffers -y", shell=True, capture_output=True)
subprocess.run("uv pip install --python .venv 'mediapipe>=0.10.0' 'protobuf>=3.20,<5.0' 'flatbuffers>=2.0'", shell=True, check=True, capture_output=True)

# 5. Configure virtual display
os.system('Xvfb :1 -screen 0 2560x1440x8 &')
os.environ['DISPLAY'] = ':1.0'

clear_output()
print("✅ Installation Complete!")
print("\nInstalled Components:")
print("  ✓ Transformers 4.46.3 (Alignment Compatible)")
print("  ✓ PyTorch 2.3.1 (Stable)")
print("  ✓ WhisperX (GPU Accelerated)")
print("  ✓ InsightFace & MediaPipe (Face Detection)")
print("  ✓ All dependencies ready")
print("\n👉 Proceed to Step 2 to launch NorCuts!")
```

## 📖 How to Use NorCuts

Once the web interface is running:

1. **Click the public URL** that appears above to open NorCuts in your browser
2. **Upload a video** or paste a YouTube URL
3. **Configure settings**:
   - Choose clip duration (30-90 seconds recommended)
   - Select AI model (Gemini, GPT-4, or Local)
   - Enable/disable face tracking and split-screen
   - Customize caption style
4. **Generate clips** - The AI will identify viral moments
5. **Download** your optimized short clips

### Tips for Best Results 💡
- Use videos with clear audio for better transcription
- Enable face tracking for interview/podcast content
- Try different caption styles to match your brand
- Export to Premiere Pro XML for advanced editing

## 🆘 Troubleshooting

**Issue**: Installation fails or takes too long  
**Solution**: Make sure you're using a GPU runtime (Runtime → Change runtime type → GPU)

**Issue**: No public URL appears  
**Solution**: Wait 2-3 minutes, then check the output. If still no URL, restart the cell.

**Issue**: Out of memory errors  
**Solution**: Use shorter videos or reduce output resolution
